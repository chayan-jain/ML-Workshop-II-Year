# Data Wrangling

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
sns.heatmap(data.isnull(),yticklabels=False,cbar=False)
```

```python
sns.boxplot(x="pclass",y='age',data=data)
```

```python
data["age"] = data.groupby("pclass").transform(lambda x: x.fillna(x.mean()))
```

```python
data.head()
```

```python
sns.countplot(x="survived",hue="cabin",data=data)
```

```python
data.drop('cabin',axis=1,inplace=True)
```

```python
data.head()
```

```python
data.dropna(inplace=True)
```

```python
sns.heatmap(data.isnull(),yticklabels=False,cbar=False)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
sex=pd.get_dummies(data['sex'],drop_first=True)
sex.head()
```

```python
embark=pd.get_dummies(data['embarked'],drop_first=True)
embark.head()
```

```python
pcl=pd.get_dummies(data['pclass'],drop_first=True)
pcl.head()
```

```python
data=pd.concat([data,sex,embark,pcl],axis=1)
```

```python
data.head()
```

```python
data.drop(['sex','embarked','name','ticket','pclass'],axis=1,inplace=True) #dropping irrelevant columns
```

```python
data.head()
```

# Train Data

```python
x=data.drop('survived',axis=1)#all the features,independent
y=data["survived"]#output to be predict as need to predict whether survived or not,dependent
```

```python
### splitting data into trainng and testing subset
```

```python
from sklearn.cross_validation import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33,random_state=None)
#random state tells the order in which the data is splitted
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
predictions=logmodel.predict(x_test)
```

# Accuracy Check

```python
#to evaluate hoe the model is working can now calculate the accuracy and also can calculate the classification report
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))#to calculate accuracy 2 by 2 matrix which tells how accurate your values are
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)
```
